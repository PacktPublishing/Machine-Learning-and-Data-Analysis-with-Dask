# Merging Dataframe

```python
import holidays
import pandas as pd
```

```python
us_holidays = holidays.UnitedStates(state='NY', 
                                    years=[2018])
```

```python
holidays_df = pd.DataFrame(
        list(us_holidays.items()),
        columns=["date", "holiday"])
holidays_df["date"] = holidays_df["date"].astype("M8")
```

```python
holidays_df
```

```python
from dask.distributed import Client, LocalCluster

cluster = LocalCluster()
client = Client(cluster)
client
```

```python
import dask.dataframe as dd

path = '../Chapter2/nyctaxi/part_parquet_stage1/year=2018/'
ddf = dd.read_parquet(path)
```

```python
dt_cols = ['tpep_pickup_datetime',
           'tpep_dropoff_datetime']
```

```python
for col in dt_cols:
    ddf[col] = dd.to_datetime(ddf[col],
                             format='%m/%d/%Y %I:%M:%S %p') 
```

```python
ddf["pickup_date"] = (ddf["tpep_pickup_datetime"].dt.date)\
                     .astype("M8")
```

```python
ddf1 = dd.merge(ddf, holidays_df, 
                left_on="pickup_date", 
                right_on="date",
                how="left")
```

```python
%%time
path_out = "../Chapter2/nyctaxi/2018_trips_holidays/"
ddf1.to_parquet(path_out,
                write_index=False,
                schema="infer")
```
