```python
def extract(year):
    if year <= 2012:
        url = f"http://www.sii.cl/pagina/valores/uf/uf{year}.htm"
    else:
        url = f"http://www.sii.cl/valores_y_fechas/uf/uf{year}.htm"
    df = pd.read_html(url, thousands=".", decimal=",")
    df = df[-1]
    return df


def transform(df, year):
    cols_dict = {'Día': 'day',
      'Ene': 1, 'Feb': 2, 'Mar': 3,
      'Abr': 4, 'May': 5, 'Jun': 6,
      'Jul': 7, 'Ago': 8, 'Sep': 9,
      'Oct': 10, 'Nov': 11, 'Dic': 12}
    
    df = df.rename(columns=cols_dict)
    df = pd.melt(df,
                 id_vars="day",
                 var_name="month",
                 value_name="price")
    df = df.dropna(subset=["price"]).reset_index(drop=True)
    df["year"] = year
    df["date"] = pd.to_datetime(df[["year",
                                    "month",
                                    "day"]])
    return df


def load(df, year, fldr_out):
    os.makedirs(fldr_out, exist_ok=True)
    df.to_csv(os.path.join(fldr_out, f"{year}.csv"),
              index=False)


def ETL(year, fldr_out):
    df = extract(year)
    df = transform(df, year)
    load(df, year, fldr_out)
```
